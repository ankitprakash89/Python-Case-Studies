```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#•	What is the major reason people being kidnapped in each state.
# to read the file
kidnap = pd.read_csv("C:/Users/Administrator/Desktop/Data Visualisation/Data Set/rajanand-crime-in-india/39_Specific_purpose_of_kidnapping_and_abduction.csv", header = 0)

# sub-setting the data as per requirement
kidnap = kidnap[0:3230][['Area_Name','Year','Group_Name','K_A_Grand_Total']]

# renaming the data
kidnap.columns = (['State', 'Year', 'Reason', 'Total'])

# replacing "Kidnap - for" from whole column
kidnap['Reason'] = kidnap.Reason.replace("Kidnap - For ",'', regex=True)

# for dropping rows with "NA" values
kidnap.dropna()
k1 = kidnap.groupby(['State','Reason'])['Total'].sum().reset_index()

# plotting the heat map to demontrate the major reason for people kidnapped statewise 
df2 = k1.pivot("State", "Reason", "Total")
sns.heatmap(df2, linewidths=.3, center=1)
```
